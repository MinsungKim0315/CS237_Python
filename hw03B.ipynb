## Problem 1:  (Continuous Probability, 5 pts) 

This problem is a classic probability problem called `Buffon's Needle.`  The experiment
involves a large number of needles 1 inch long, and a large piece of paper with parallel lines
drawn 1 inch apart. The needles are dropped onto the paper randomly, and we count the number
of needles which cross one of the lines.  

The analytical probability  that a needle crosses a line is

$$2/\pi \ =\ 0.6366197723675814...$$

(see Wikipedia for a derivation). 

In this problem, we will simulate this experiment, and use it to calculate $\pi$. Such methods of
calculation involving random trials are called "Monte Carlo" algorithms, in honor of the famous
casino, where gamblers to this day still show off their ignorance of probability.

```python
# (A) Complete the template and run with 10 trials and using seed(0)

# Experiment: comment out the seed(0) and run this cell several times to see the distribution of points,
# with num_trials = 10, 100, etc.  

import random
num_trials = 100
seed(0)
def random_line():
    global theta
    x1 = random.uniform(0, 10)
    y1 = random.uniform(0, 10)
    theta = random.uniform(0, 2*math.pi)
    x2 = x1 + math.cos(theta)
    y2 = y1 + math.sin(theta)
    return ((x1,y1), (x2, y2))
random_line()
#-------------------------------------------------------------------------------------------------------------
def random_needle_plot(num_trials,display=True):
    print("\nRunning simulation with",f"{num_trials:,}","trials....")
    
    # For displaying the simulation 
    if(display):
        plt.figure(num=None, figsize=(10, 10))
        plt.title('Buffon\'s Needle Simulation with '+ str(num_trials)+ ' trials.',fontsize=12)
        for k in range(11):
            plt.plot([k,k ],[ -1,11],color='k',alpha=0.25)     # draw grid lines  
        plt.ylim(-1, 11)
        plt.xlim(-1, 11)

    count = 0
    for k in range(num_trials):
        
        ((x1,y1),(x2,y2)) = random_line()
        ceil = math.ceil((x1+x2)/2)  
        if((ceil - ((x1+x2)/2)) < 0.5):
            distance = ceil - (x1+x2)/2
        elif((ceil - ((x1+x2)/2)) > 0.5):
            distance = 1 - (ceil - ((x1+x2)/2))
        else:
            pass
        
        if(distance < abs(math.cos(theta)/2)):
            count = count + 1
            
        if(display):
            plt.plot([x1,x2],[y1,y2])    # draw the line
           
    if(display):
        plt.show()
        
    return(count)
    
random_needle_plot(num_trials)    
```
